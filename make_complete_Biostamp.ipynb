# This notebook creates the Biostamp_complete.csv ground truth file

```python
import pandas as pd
import numpy as np
import os
```

## Get aggregate feature files

```python
agfiles = os.listdir("AGData/")
agfiles.sort()
agfiles
```

## Get ground truth files

```python
gtfiles = os.listdir("GroundTruth_new/")
gtfiles.sort()
gtfiles
```

## Add column with activity domain type

```python
types = ["L-1", "H-1", "H-2", "A-2", "L-3", "H-3", "W-4", "H-4", "L-5", "E-6", "A-6", "H-7", "E-7", "E-8", "E-8",
        "A-8", "W-9", "A-9", "W-10", "L-11", "A-11", "A-11", "A-12", "L-12", "L-12", "W-13", "A-13", "W-14", "E-14",
        "W-15", "E-15", "E-15", "L-16", "W-16", "E-17", "W-17", "E-18", "L-18", "H-19", "W-19", "H-20", "E-20",
        "A-21", "W-21", "E-22", "L-22", "L-24", "A-24", "A-24", "H-25", "L-25", "A-26", "H-26", "H-26", "E-27", "H-27"]
```

```python
for i in range(0, len(gtfiles)):
    sec_codings = pd.read_csv('GroundTruth_new/' + gtfiles[i])
    type_array = np.full(sec_codings.shape[0], types[i])
    print(type_array)
    sec_codings['type'] = type_array
    sec_codings.to_csv(gtfiles[i])
    print(gtfiles[i])
```

```python
copyBobs = ["AM02DO2", "AM08DO1", "AM11DO1", "AM11DO2", "AM12DO2", "AM15DO2", "AM26DO2"]
```

## Generate complete Biostamp ground truth file

```python
complete = pd.DataFrame()

for i in range(0, len(gtfiles)):
    gt = pd.read_csv(gtfiles[i])

    pid = gtfiles[i][:4]
    session = gtfiles[i][4:7]

    if (gtfiles[i][:7] not in copyBobs):
        features = pd.read_csv("AGData/" + pid + "-" + session + ".csv")
    else:
        features = pd.read_csv("AGData/" + pid + "-" + session + "_" + gtfiles[i][14] + ".csv")
    
    merge = pd.merge(features, gt, on='Unnamed: 0')
    complete = complete.append(merge)
    print("Merged " + pid + "-" + session)
```

```python
list(complete)
```

```python
complete = complete.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y', 'Unnamed: 0.1', 'Unnamed: 0.1.1', 'observation', 'date', 'time'])
```

```python
complete.to_csv('new_Biostamp_complete.csv')
```

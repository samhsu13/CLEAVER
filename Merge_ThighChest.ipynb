# This notebook merges the thigh and chest aggregate data generated by Features.R

```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
mydir = "/Users/samhsu/Documents/THESIS/Data/BioStampOut/"
files = os.listdir(mydir)
files.sort()
files = files[1:]
files
```

```python
for i in range(0, len(files), 2):
    chest = pd.read_csv(mydir + files[i])
    thigh = pd.read_csv(mydir + files[i+1])
    merged = pd.merge(thigh, chest, on='start.time')
    merged.to_csv(files[i][:-9] + ".csv")
    print("merged " + files[i][:-9] + ".csv")
```

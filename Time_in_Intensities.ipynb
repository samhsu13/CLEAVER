Create csv for total time spent sedentary, light, mod and vig for each subject/observation

```python
import pandas as pd
import numpy as np
```

```python
my_dir = "/Users/samhsu/Documents/THESIS/Out/"
ground_truth_log = pd.read_csv("/Users/samhsu/Documents/THESIS/Data/best_obs_onesheet.csv")
```

```python
def read_ground_truth(ob):
    file = ob + "_second_codings.csv"
    ground_truth = pd.read_csv(my_dir + file)
    return ground_truth
```

```python
# Get the names of the observations and remove the observations we excluded
observations = set(ground_truth_log['Observation'])
observations.remove("AM10DO1_R_FINAL_C")
observations.remove("AM02DO2_J_copyB_FINAL_R") 
observations.remove("AM11DO1_N_copyB_FINAL_C") 
observations.remove("AM15DO2_M_copyB_FINAL_C") 
observations.remove("AM15DO2_N_copyB_FINAL")
observations
```

```python
def build_intensities():
    intensities = pd.DataFrame()

    for obs in observations:
        ground_truth = read_ground_truth(obs)
        ground_truth = ground_truth[:-1]
        ob_intensities = ground_truth['primary_intensity'].value_counts().index.tolist()

        sed = light = mod = vig = 0

        if ('sedentary' in ob_intensities):
            sed = ground_truth['primary_intensity'].value_counts()['sedentary']
        if ('light' in ob_intensities):
            light = ground_truth['primary_intensity'].value_counts()['light']
        if ('moderate' in ob_intensities):
            mod = ground_truth['primary_intensity'].value_counts()['moderate']
        if ('vigorous' in ob_intensities):
            vig = ground_truth['primary_intensity'].value_counts()['vigorous']

        temp = pd.DataFrame({'Observation':[obs],
                          'Sedentary':[sed],
                          'Light':[light],
                          'Moderate':[mod],
                          'Vigorous':[vig]})
        
        intensities = intensities.append(temp, ignore_index=True)
        
    cols = intensities.columns.tolist()
    cols = cols[2:4] + cols[0:2] + cols[4:]
    intensities = intensities[cols]
    
    return intensities
```

```python
intensities = build_intensities()
intensities = intensities.sort_values(by=['Observation'])
intensities
```

```python
intensities.to_csv('Intensities.csv')
```
